```python
from LifeBuddyWebApp import create_app, db
from LifeBuddyWebApp.models import User, Post, Variables
proxyApp=create_app()
ctx=proxyApp.app_context()
ctx.push()
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit, leastsq
import matplotlib.pyplot as plt
```

```python
y = np.array([12, 8, 11, 7, 5, 2, 3, 5, 6, 4, 5, 7, 8, 13, 19, 22, 25])
x = np.array(range(len(y)))
```

```python
def func1(params, x, y):
    a, b, c = params[0], params[1], params[2]
    residual = y-(a*x**2+b*x+c)
    return residual
```

```python
params=[0, 0, 0]
```

```python
result = leastsq(func1, params, (x, y))
a, b, c = result[0][0], result[0][1], result[0][2]
yfit1 = a*x**2+b*x+c

plt.plot(x, y, 'bo', label="y-original")
plt.plot(x, yfit1, color="red", label="y=ax^2+bx+c")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.grid(True)
plt.show()
```
